```python
import numpy as np
def requirement_generator(actors: List[str], activities: List[str], objects: List[str], conditions: List[str])-> str:
    modal_verbs = ["shall", "should", "must", "need to", "may", "might"]

    # actor =
    # modal = 
    # activity = 
    # object = 
    # condition = 
    

    req = f"The { np.random.choice(actors)} {np.random.choice(modal_verbs)} able to {np.random.choice(activities)} by it's {np.random.choice(objects)} {np.random.choice(conditions)}"

    return req
```

```python
df_actor = pd.read_csv("actor.csv")
actors = df_actor["actor"].values.tolist()

df_verbs = pd.read_csv("rt.verbs.csv")
activities = df_verbs["verbs"].values.tolist()
```

```python
from tqdm import tqdm

requirement_collection: List[str] = []
for item in tqdm(range(10000)):
    data = requirement_generator(actors=actors, activities=activities, objects=["photo", "videos"], conditions=["in under 30 seconds", "successfully", "quicks"])
    requirement_collection.append(data)
```
